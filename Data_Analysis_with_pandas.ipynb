```python
#8. Create a new attribute with name Premium[Sold price - Base price]
ipl_df['Premium'] = ipl_df['SOLD PRICE'] - ipl_df['BASE PRICE']
print(ipl_df)
```

```python
#9. Display five observations in descending order with respect to sold price
ipl_df[['PLAYER NAME','SOLD PRICE']].sort_values('SOLD PRICE', ascending = False)[0:5]
```

```python
#10. Create new attribute soldprice by age[mean of age wise sold price attributes]
Soldprice_by_age = ipl_df.groupby('AGE')['SOLD PRICE'].mean().reset_index()
print(Soldprice_by_age)
```

```python
#11. Create frequency distribution plot of discrete attribute[Soldprice_by_age] using bar plot
import seaborn as sn
sn.barplot(x = 'AGE', y = 'SOLD PRICE', data = Soldprice_by_age);
```

```python
#12. Create a frequency distribution plot of a continuous attribute using Histogram[Sold price attribute]
import matplotlib.pyplot as plt
plt.hist(ipl_df['SOLD PRICE'], bins=20)
```

```python
#13. Find relationship between two attributes Sixers and Sold price by scatter plot
plt.scatter(x = ipl_df['SIXERS'], y = ipl_df['SOLD PRICE'])
plt.xlabel('SIXERS')
plt.ylabel('SOLD PRICE')
plt.title('Scatter plot showing relationship')
```

```python
#14. Repeat above qustion with inclusion of direction of relationship
sn.regplot(x=ipl_df['SIXERS'], y=ipl_df['SOLD PRICE'])
```

```python
#15. Create an array of attributes called as Influential attributes and prepare pair plot
influential_features = ['SR-B','AVE','SIXERS','SOLD PRICE']
sn.pairplot(ipl_df[influential_features], height = 2)
```

```python
#16. Find the correlation of influential feature array
ipl_df[influential_features].corr()
```

```python
#17. Highlight the correlation value with heatmap function
sn.heatmap(ipl_df[influential_features].corr(), annot = True)
```

```python
#18. Prepare a boxplot to calculate range of attribute soldprice
box = sn.boxplot(ipl_df['SOLD PRICE'])
```

```python
#19. Repeat the above problem with matplot library
plt.boxplot(ipl_df['SOLD PRICE'])
```

```python
#20. Find maximum and minimum value of sold price attribute
import numpy as np
soldprice_arr = ipl_df['SOLD PRICE']
print(np.min(soldprice_arr))
print(np.max(soldprice_arr))
```

```python
#21. Find IQR valur of sold price
Q1 = np.percentile(soldprice_arr, 25)
Q3 = np.percentile(soldprice_arr, 75)
IQR = Q3-Q1
print(IQR)
```

```python
#22. Find the median of Sold price
Q2 = np.percentile(soldprice_arr, 50)
Median = Q2
print(Median)
```
